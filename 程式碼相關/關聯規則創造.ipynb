```python
OrderSlaveData = pd.read_csv("D:/91APP_DataSet_2023/91APP_DataSet_2023/OrderSlaveData.csv")
OrderSlaveData.head()
```

```python
file = pd.read_csv("C:/Users/user/Desktop/bda/multihot_vector.csv")
item_features = dict()
for _, row in file.iterrows():
    r = -2
    feature = []
    for j in row:
        r += 1
        if j == 1:
            feature.append(int(r))
    item_features[row["SalePage_Id"]] = feature
```

```python
OrderSlaveData = OrderSlaveData.dropna()
```

```python
# tradegroup_slave = OrderSlaveData["TradesGroupCode"].unique()
# print(len(tradegroup_slave))

cart_base = dict()
people_base = dict()
from tqdm import tqdm, trange
for _, row in tqdm(OrderSlaveData.iterrows()):
    if int(row["OrderDateTime"][0:4]) >= 2023 or int(row["OrderDateTime"][0:4]) <= 2020:
        continue
    if row["TradesGroupCode"] not in cart_base:
        cart_base[row["TradesGroupCode"]] = set(item_features[row["SalePageId"]])
    else:
        set.union(cart_base[row["TradesGroupCode"]],set(item_features[row["SalePageId"]]))
    if row["ShopMemberId"] not in people_base:
        people_base[row["ShopMemberId"]] = set(item_features[row["SalePageId"]])
    else:
        set.union(people_base[row["ShopMemberId"]], set(item_features[row["SalePageId"]]))
print(cart_base)
print(people_base)
```

```python
# cart_base
basket = list(cart_base.values())
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
te = TransactionEncoder()
te_ary = te.fit_transform(basket)
df = pd.DataFrame(te_ary, columns=te.columns_)
print(df)
# 使用 Apriori 演算法找出頻繁項目集
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=False)
print(frequent_itemsets)

# 基於頻繁項目集產生關聯規則
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# 輸出結果
print("頻繁項目集:")
print(frequent_itemsets)
print("\n關聯規則:")
print(rules)
```

```python
# people_base
basket = list(people_base.values())
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
te = TransactionEncoder()
te_ary = te.fit_transform(basket)
df = pd.DataFrame(te_ary, columns=te.columns_)
print(df)
# 使用 Apriori 演算法找出頻繁項目集
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=False)
print(frequent_itemsets)

# 基於頻繁項目集產生關聯規則
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# 輸出結果
print("頻繁項目集:")
print(frequent_itemsets)
print("\n關聯規則:")
print(rules)
```
