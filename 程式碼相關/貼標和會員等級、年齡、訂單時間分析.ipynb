```python
def age_index(age):
  df["age_group"] = "nan"
  df.loc[(age <= 20), "age_group"] = "<=20"
  df.loc[(age > 20) & (age <= 30), "age_group"] = "20-30"
  df.loc[(age > 30)& (age <= 40), "age_group"] = "30-40"
  df.loc[(age > 40)& (age <= 50), "age_group"] = "40-50"
  df.loc[(age >= 50), "age_group"] = ">=50"

age_index(df["age"].values)
df = df.loc[df["age"]<=80]
df = df.loc[df["age"]>=10]
df
```

```python
def level_index(memberlevel):
  df["level_group"] = "nan"
  df.loc[(memberlevel == 10), "level_group"] = "low"
  df.loc[(memberlevel == 20), "level_group"] = "mid"
  df.loc[(memberlevel == 30), "level_group"] = "hig"

level_index(df["MemberCardLevel"].values)
df = df.loc[df["age"]<=80]
df = df.loc[df["age"]>=10]
df
```

```python
def level_age_index(age, memberlevel):
  df["mix_index"] = "nan"
  lst = ["low", "mid", "hig"]
  for i in range(1,4):
    df.loc[(memberlevel == i * 10) & (age <= 20), "mix_index"] = "<=20" + lst[i-1]
    df.loc[(memberlevel == i * 10) & (age > 20) & (age <= 30), "mix_index"] = "20-30"+ lst[i-1]
    df.loc[(memberlevel == i * 10) & (age > 30) & (age <= 40), "mix_index"] = "30-40"+ lst[i-1]
    df.loc[(memberlevel == i * 10) & (age > 40) & (age <= 50), "mix_index"] = "40-50"+ lst[i-1]
    df.loc[(memberlevel == i * 10) & (age > 50),  "mix_index"] = ">=50"+ lst[i-1]

level_age_index(df["age"].values, df["MemberCardLevel"].values)
df = df.loc[df["age"]<=80]
df = df.loc[df["age"]>=10]
df
```

```python
def group_analysis(dict_group, df):
    for key in dict_group:
      data = dict_group[key]
      list1 = data[["TotalSalesAmount","TotalPrice", "TotalDiscount", "TotalPromotionDiscount", "TotalCouponDiscount"]].sum()
      discount_ratio_mean = data[data["discount_ratio"]!=0]["discount_ratio"].mean().round(2)
      discount_ratio_std = data[data["discount_ratio"]!=0]["discount_ratio"].std().round(2)
      list1["average_spending"] = ((list1["TotalSalesAmount"] / len(data))).round(1)
      list1["stddev_spending"] = (data["TotalSalesAmount"].std()).round(1)
      list1["order_num"] = len(data)
      list1["discount_ratio_stat"] = [discount_ratio_mean, discount_ratio_std]
      discount_ratio_num = len(data)
      discount_ratio_num2 = len(data[data["discount_ratio"]!=0]["discount_ratio"])
      list1["discount_percentage(%)"] = 100*(round(discount_ratio_num2 / discount_ratio_num, 3))
      #list1["Total_discount_ratio"] = 100*(round(-list1["TotalDiscount"] / list1["TotalPrice"], 3))
      #list1["Promotion_discount_ratio"] = 100*(round(-list1["TotalPromotionDiscount"] / list1["TotalPrice"], 3))
      #list1["Coupon_discount_ratio"] = 100*(round(-list1["TotalCouponDiscount"] / list1["TotalPrice"], 3))
      df[key] = list1
    df = df.transpose()
    return df
```
