```python
import pandas as pd

# 第一張表
df1 = pd.DataFrame({
    'name': ['白衣', '黑衣', '白裙', '黃褲', '紅衣'],
    'sale': [300, 400, 333, 666,234]
})

# 第二張表
df2 = pd.DataFrame({
    'item': ['衣', '裙', '褲'],
    'label': [1, 2, 3]
})

# 合併表格
result_df = pd.DataFrame(columns=['name', 'sale', 'title'])

for index, row in df2.iterrows():
    print(df1['name'])
    print(row['item'])
    print(df1['name'].str.contains(row['item']))
    print("====")
    match_df = df1[df1['name'].str.contains(row['item'])].copy()  # Make a copy of the slice
    match_df.loc[:, 'title'] = row['label']  # Use .loc to set value on the copy
    result_df = pd.concat([result_df, match_df], ignore_index=True)

print(result_df)
```

```python
import pandas as pd
RuleData = pd.read_csv("D:/91APP_DataSet_2023/91APP_DataSet_2023/rules.csv")

# 合併表格
result_df = pd.DataFrame(columns=['label'])

SalePageData = SalePageData.dropna()
SalePageData = SalePageData.apply(lambda x: x.astype(str).str.lower())


for index, row in RuleData.iterrows():
    match_df = SalePageData[SalePageData['SalePage_Title'].str.contains(row['item'])].copy()  # Make a copy of the slice
    match_df.loc[:, 'label'] = row['label']  # Use .loc to set value on the copy
    result_df = pd.concat([result_df, match_df], ignore_index=True)

result_df.columns = ['label','SalePageId','SalePage_Title']
print(result_df)
result_df.to_csv('result_df.csv', index=False)
```

```python
data  = merged_df[0:10000].groupby('TradesGroupCode')['SalePageId'].unique().apply(list).tolist()
print(data)
```

```python
from apyori import apriori
association_rules = apriori(data, min_support=0.16, min_confidence=0.2, min_lift=3, max_length=2) 
association_results = list(association_rules)
print("!!!")
for item in association_results:
   pair = item[0] 
   items = [x for x in pair]
   print("Rule: " + items[0] + " -> " + items[1])
   print("Support: " + str(item[1]))
   print("Confidence: " + str(item[2][0][2]))
   print("Lift: " + str(item[2][0][3]))
   print("=====================================")
```
